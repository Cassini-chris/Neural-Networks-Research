```python
import tensorflow as tf
from tensorflow import keras
from keras import backend as K
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

x_zoomed_in = np.linspace(-10,10,100)
x = np.linspace(-100,100,100)
```

```python
def heartbeat_activation(x):
      return (K.maximum(x*0,K.sin(x))/x* ((abs(x+8)) /((13*(abs(-1))**4) /7)+1))

plt.plot(x_zoomed_in, heartbeat_activation(x_zoomed_in), label = 'Heartbeat')
plt.legend(fontsize = 12)
plt.xlabel('x')
plt.show()

plt.plot(x, heartbeat_activation(x), label = 'Heartbeat')
plt.legend(fontsize = 12)
plt.xlabel('x')
plt.show()
```
